```python
import cmdstanpy
import arviz as az
import xarray as xr
```

```python
draws2data_model = cmdstanpy.CmdStanModel(stan_file="/Users/dashadower/git_repos/stanify/vignette/stan_files/hier_prey_predator_wopnoise/draws2data_hier_prey_predator_wopnoise.stan")
```

```python
draws2data_fit = draws2data_model.sample(chains=1, iter_sampling=1000, fixed_param=True)
```

```python
data_name = "prey_obs"
```

```python
draws2data_idata = az.from_cmdstanpy(prior=draws2data_fit, prior_predictive=[data_name])
draws2data_idata
```

```python
draws2data_idata = draws2data_idata.stack(prior_draw=["chain", "draw"], groups="prior_groups", create_index = False)
```

```python
zeroth_data = draws2data_idata.isel(prior_draw=0)
zeroth_data.prior_predictive[data_name]
```

```python
data2draws_model = cmdstanpy.CmdStanModel(stan_file="/Users/dashadower/git_repos/stanify/vignette/stan_files/prey_predator_wopnoise/data2draws_prey_predator_wopnoise.stan")
```

```python
data2draws_fit = data2draws_model.sample(data={"prey_obs": zeroth_data.prior_predictive[data_name]}, iter_sampling=100)
```

```python
data2draws_idata = az.from_cmdstanpy(posterior=data2draws_fit)
```

```python
data2draws_idata
```

```python
data2draws_idata.stack(posterior_draw=["chain", "draw"], groups="posterior_groups", create_index = False)
```

```python
post = xr.concat((data2draws_idata.posterior, ), dim="prior_draw")
```

```python
post
```

```python
draws2data_idata.add_groups(posterior=post)
```
